```python
import pandas as pd
from scipy.stats import chi2_contingency

df = pd.read_csv('watch_data (5).csv')

luxury_brands = ['Rolex', 'Audemars Piguet', 'Patek Philippe', 'Tudor', 'Hublot', 'Vacheron Constantin']
mid_range_brands = ['Omega', 'Breitling', 'Zenith', 'Cartier', 'TAG Heuer', 'IWC', 'Longines']

def categorize_brand(brand):
    if brand in luxury_brands:
        return 'Luxury'
    elif brand in mid_range_brands:
        return 'Mid-Range'
    else:
        return 'Non-Luxury'

df['Watch_Category'] = df['Brand'].apply(categorize_brand)

print(df['Watch_Category'].value_counts())

lunette_counts = df.groupby(['Watch_Category', 'Lunette Material']).size().unstack(fill_value=0)

total_luxury = lunette_counts.loc['Luxury'].sum()
total_mid_range = lunette_counts.loc['Mid-Range'].sum()
total_non_luxury = lunette_counts.loc['Non-Luxury'].sum()

prob_luxury = lunette_counts.loc['Luxury'] / total_luxury
prob_mid_range = lunette_counts.loc['Mid-Range'] / total_mid_range
prob_non_luxury = lunette_counts.loc['Non-Luxury'] / total_non_luxury

prob_lunette = pd.DataFrame({
    'Luxury': prob_luxury,
    'Mid-Range': prob_mid_range,
    'Non-Luxury': prob_non_luxury
}).fillna(0)

print(prob_lunette)

contingency_table_lunette = lunette_counts.T
chi2_lunette, p_lunette, dof_lunette, expected_lunette = chi2_contingency(contingency_table_lunette)

print(f"Lunette Material - Chi-square value: {chi2_lunette}, p-value: {p_lunette}")
print(expected_lunette)

case_counts = df.groupby(['Watch_Category', 'Case material']).size().unstack(fill_value=0)

total_luxury = case_counts.loc['Luxury'].sum()
total_mid_range = case_counts.loc['Mid-Range'].sum()
total_non_luxury = case_counts.loc['Non-Luxury'].sum()

prob_luxury = case_counts.loc['Luxury'] / total_luxury
prob_mid_range = case_counts.loc['Mid-Range'] / total_mid_range
prob_non_luxury = case_counts.loc['Non-Luxury'] / total_non_luxury

prob_case = pd.DataFrame({
    'Luxury': prob_luxury,
    'Mid-Range': prob_mid_range,
    'Non-Luxury': prob_non_luxury
}).fillna(0)

print(prob_case)

contingency_table_case = case_counts.T
chi2_case, p_case, dof_case, expected_case = chi2_contingency(contingency_table_case)

print(f"Case Material - Chi-square value: {chi2_case}, p-value: {p_case}")
print(expected_case)

bracelet_counts = df.groupby(['Watch_Category', 'Bracelet material']).size().unstack(fill_value=0)

total_luxury = bracelet_counts.loc['Luxury'].sum()
total_mid_range = bracelet_counts.loc['Mid-Range'].sum()
total_non_luxury = bracelet_counts.loc['Non-Luxury'].sum()

prob_luxury = bracelet_counts.loc['Luxury'] / total_luxury
prob_mid_range = bracelet_counts.loc['Mid-Range'] / total_mid_range
prob_non_luxury = bracelet_counts.loc['Non-Luxury'] / total_non_luxury

prob_bracelet = pd.DataFrame({
    'Luxury': prob_luxury,
    'Mid-Range': prob_mid_range,
    'Non-Luxury': prob_non_luxury
}).fillna(0)

print(prob_bracelet)

contingency_table_bracelet = bracelet_counts.T
chi2_bracelet, p_bracelet, dof_bracelet, expected_bracelet = chi2_contingency(contingency_table_bracelet)

print(f"Bracelet Material - Chi-square value: {chi2_bracelet}, p-value: {p_bracelet}")
print(expected_bracelet)

complications_counts = df.groupby(['Watch_Category', 'Complication']).size().unstack(fill_value=0)

total_luxury = complications_counts.loc['Luxury'].sum()
total_mid_range = complications_counts.loc['Mid-Range'].sum()
total_non_luxury = complications_counts.loc['Non-Luxury'].sum()

prob_luxury = complications_counts.loc['Luxury'] / total_luxury
prob_mid_range = complications_counts.loc['Mid-Range'] / total_mid_range
prob_non_luxury = complications_counts.loc['Non-Luxury'] / total_non_luxury

prob_complications = pd.DataFrame({
    'Luxury': prob_luxury,
    'Mid-Range': prob_mid_range,
    'Non-Luxury': prob_non_luxury
}).fillna(0)

print(prob_complications)

contingency_table_complications = complications_counts.T
chi2_complications, p_complications, dof_complications, expected_complications = chi2_contingency(contingency_table_complications)

print(f"Complications - Chi-square value: {chi2_complications}, p-value: {p_complications}")
print(expected_complications)

contingency_table = pd.crosstab(df['Watch_Category'], df['Complication'])
print(contingency_table)

chi2_value, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square value: {chi2_value}")
print(f"p-value: {p_value}")
print(expected)

df.to_csv('watch_data (5).csv', index=False)
files.download('watch_data (5).csv')
```
